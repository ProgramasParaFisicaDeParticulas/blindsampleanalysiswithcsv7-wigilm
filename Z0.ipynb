```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go


data=pd.read_csv('https://raw.githubusercontent.com/ProgramasParaFisicaDeParticulas/blindsampleanalysiswithcsv7-wigilm/main/Sample7.csv')
print(len(data))
data.head(10)
```

Veamos como se comportan los detectores viendo las graficas de los ángulos phi y eta

```python
fig_test=px.histogram(data.phi1)
fig_test.show()
fig_test2=px.histogram(data.phi2)
fig_test2.show()
fig_test3=px.scatter(x=data.phi1,y=data.eta1)
fig_test3.show()
fig_test4=px.scatter(x=data.phi2,y=data.eta2)
fig_test4.show()
```

```python
def minvariant(pt1,pt2,eta1,eta2,phi1,phi2):
  return np.sqrt(2*pt1*pt2*(np.cosh(eta2-eta1)-np.cos(phi1-phi2)))
minv=minvariant(data['pt1'],data['pt2'],data['eta1'],data['eta2'],data['phi1'],data['phi2'])

fig_1=px.histogram(minv)
fig_1.show()

fig_2=px.histogram(data['Q1']+data['Q2'])
fig_2.show()
```

La masa invariante de los datos crudos recuerda a aquella del Z0. Parece que nuevamente los eventos están entre 60Gev/c^2 y 120 Gev/c^2. Sin embargo, ahora hay muchos eventos con carga total distinta de 0. Veamos que pasa si se quitan

```python
data1=data[data.Q1+data.Q2==0]
minv1=minvariant(data1['pt1'],data1['pt2'],data1['eta1'],data1['eta2'],data1['phi1'],data1['phi2'])
print(len(data1))

fig_3=go.Figure()
fig_3.add_trace(go.Histogram(x=minv, name='eventos raw',nbinsx=200))
fig_3.add_trace(go.Histogram(x=minv1, name='eventos Q=0',nbinsx=200))
fig_3.update_layout(barmode='overlay')
fig_3.show()
```

Vemos que ahora solamente quedan 7591 datos de los 10,000 con los que empezamos. Además, la gráfica se ha aplanado más del pico, se esperaría que el pico fuera más pronunciado después de haber eliminado estos datos. Revisemos si son electrones o muones los datos para ver si se encuentra algo interesante.

```python
tol=0.02
bins=1000
data_jets=data1[(data1.sigmaEtaEta1 <=tol )  & (data1.sigmaEtaEta2 <=tol)]
print(len(data_jets))
minvjets=minvariant(data_jets['pt1'],data_jets['pt2'],data_jets['eta1'],data_jets['eta2'],data_jets['phi1'],data_jets['phi2'])

fig_4=go.Figure()
fig_4.add_trace(go.Histogram(x=minv, name='eventos raw'))
fig_4.add_trace(go.Histogram(x=minv1, name='eventos Q=0'))
fig_4.add_trace(go.Histogram(x=minvjets, name='eventos Q=0 y sigmaetaeta<=tol'))
fig_4.update_layout(barmode='overlay')
fig_4.show()
```

Ahora vamos a ver la calidad de los electrones discriminando los datos que solo se detectan en un detector, empezemos con los datos originales y pongamosle un filtro de detector EE
